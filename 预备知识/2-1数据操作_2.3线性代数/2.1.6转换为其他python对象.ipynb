将深度学习框架定义的张量转换为NumPy张量（ndarray)非常容易，只需调用张量对象的numpy方法即可。反之也然，我们可以通过传入一个NumPy张量来创建一个张量对象，深度学习框架会自动将其转换为张量

torch张量和NumPy张量共享内存空间，因此更改其中一个会影响另一个

```python
x = torch.arange(12, dtype = torch.float32).reshape((3, 4))
a = x.numpy()
type(a)
```

```python
b = torch.tensor(a)
type(b)
```

要将大小为1的张量转换为Python标量，可以调用item方法或使用Python的内置函数

```python
A = torch.tensor([3.5])
A, A.item(), float(A), int(A)
```

深度学习储存和操作数据的主要接口是张量（tensor)。张量是一个多维数组，类似于NumPy中的ndarray。与NumPy不同的是，深度学习框架的张量可以在GPU上运行，从而加快计算速度

他提供了各种功能，包括基本数学运算，广播，索引，切片，内存节省和转换为其他对象等。
