按照元素运算是张量操作中最常见的操作。假设我们有两个形状相同的张量，我们可以对它们进行四则运算

```python
import torch
```

```python
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
```

```python
x + y, x - y, x * y, x / y
```

还可以求幂

```python
torch.exp(x)
```

```python
x
```

还可以把多个张量连接到一起，把它们端对端地地欸来形成一个更大的张量，需要提供张量列表，并给出沿哪个轴连接·

```python
 x = torch.arange(12, dtype = torch.float32).reshape((3, 4))
 y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
h = torch.cat((x, y), dim = 0), torch.cat((x, y), dim = 1)
x, y, h
```

有时我们想通过逻辑运算符构建二元张量，这时可以使用比较运算符

```python
x == y
```
