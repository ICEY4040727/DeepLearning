（1）通过适当复制元素来拓展较小的张量，使其形状与较大张量相同。

（2）对生成的数组执行按元素运算。

大多数情况下我们将沿着数组中长度为1的轴进行广播。# # % % md

```python
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

如下，a的左列复制到右列，b的上行复制到下行，然后再按元素相加
