```python
import torch
```

```python
x = torch.arange(12)
```

```python
x
```

通过张量的形状，我们可以知道它有多少个轴（维度）以及每个轴上有多少个元素。

```python
x.shape
```

想知道张量有多少个元素，可以使用 numel 方法

```python
x.numel()
```

,这是因为处理的是单向量，所以他的形状和元素书相同。

如果我们想要一个矩阵，我们可以调整张量的形状。想要改变一个张量的形状而不改变其数据，可以使用 reshape 方法

```python
x = x.reshape(3, 4)
```

```python
x
```

好，我们先回复x的样子

```python
x = x.reshape(12)
x
```

通过指明一个维度的大小为-1，可以让PyTorch根据其他维度的大小推断出该维度的大小

```python
y = x.reshape(2, -1, 3)
y
```

同样，我们有一些能够创建特定形状张量的函数。例如，下面的代码创建了一个3行4列的矩阵，矩阵中的所有元素都初始化为0

```python
torch.zeros((3, 4))
```

或者全都初始化为1

```python
torch.ones (2, 3, 4)
```

或者随机

```python
torch.randn(3, 4)
```

```python
torch.tensor([[1, 2, 3], [4, 5, 6]])
```
