执行一些操作可能为结果分配新的内存空间，而有些操作则会重用原有的内存空间。

例如，下面的代码创建了一个张量x，然后将其赋值给y。此时x和y指向同一块内存空间

```python
x = torch.arange(12, dtype = torch.float32).reshape((3, 4))
x
y = x
id(x) == id(y)
```

```python
而相加操作会为结果分配新的内存空间，因此此时x和y指向不同的内存空间
```

```python
y = x + y
id(x) == id(y)
```

在机器学习中，我们可能有巨大数量的参数，因此我们希望尽可能节省内存空间。为此，我们可以使用原地（in-place）操作来避免不必要的内存分配和数据复制。

以及，如果我们不原地更新，一些引用了旧值的变量将变得过时，这样我们的代码无意间会使用过时的值，导致难以发现的错误。

```python
我们可以使用切片表示法来原地更改张量的某些元素的值。
```

```python
z = torch.zeros_like(y)
z
```

```python
id(z)
```

```python
z[:] = x + y
z,id(z)
```
